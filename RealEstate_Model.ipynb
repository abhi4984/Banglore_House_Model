### Feature Engineering

```python
df.head()
```

```python
df['total_sqft'] = df['total_sqft'].apply(lambda x:float(x))
```

```python
df.head()
```

```python
df['Price_Per_sqft'] = df['price']*100000/df['total_sqft']
```

```python
df
```

```python
len(df.location.unique())
```

```python
loc_stat = df['location'].value_counts()
dict(loc_stat)
```

```python
len(loc_stat[loc_stat] <= 10)
```

```python
loc_stat_less_10 = loc_stat[loc_stat <= 10]
loc_stat_less_10
```

```python
df.location = df.location.apply(lambda x:'other' if x in loc_stat_less_10 else x)
```

```python
df
```

```python
df = df[~(df.total_sqft/df.BHK < 300)]
```

```python
def predict_price(location,sqft,bath,bhk):
    loc_index = np.where(x.columns==location)[0][0]
    z = np.zeros(len(x.columns))
    z[0] = sqft
    z[1] = bath
    z[2] = bhk
    if(loc_index >= 0):
        z[loc_index] = 1
    return model.predict([z])[0]
```
